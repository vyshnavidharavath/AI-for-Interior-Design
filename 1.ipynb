```python
import torch
from diffusers import StableDiffusionPipeline
import gradio as gr
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    use_safetensors = True
)

pipe = pipe.to("cpu")   # use "cuda" if GPU available
def generate_design(room_type, style, color, lighting):
    prompt = f"{room_type} interior design, {style} style, {color} color theme, {lighting} lighting, high quality, realistic"
    
    image = pipe(prompt, num_inference_steps=30, guidance_scale=7.5).images[0]
    
    return image
interface = gr.Interface(
    fn=generate_design,
    inputs=[
        gr.Dropdown(["Living Room", "Bedroom", "Kitchen"], label="Room Type"),
        gr.Dropdown(["Modern", "Minimalist", "Luxury", "Traditional"], label="Design Style"),
        gr.Dropdown(["White", "Dark", "Pastel", "Wooden"], label="Color Theme"),
        gr.Dropdown(["Warm", "Cool", "Natural", "Bright"], label="Lighting")
    ],
    outputs="image",
    title="AI Interior Design Generator",
    description="Generate interior design using AI"
)
interface.launch()
```
